### 3a) Import the necessary libraries and modules

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
random.seed(0)
```

### 3b)

<b> Student number: 0361980 </b>
<ul><li>a = 0</li><li>b = 8</li><li>c = 9</li><li>d = 1</li>

f(x) = $(a+1)x^3 + (b+1)x^2 + (c+1)x + d  $
   
f(x) = $x^3 + 9 x^2 + 10x + 1$

```python
# Our original polynomial is x^3+9*x^2+10x+1
coefs = [1,10,9,1]

def polynomial(coefs, x):
    n = len(coefs)
    return sum([coefs[i]*x**i for i in range(n)])

def draw_polynomial(coefs):
    n = len(coefs)
    x = np.linspace(-5, 5, 1000)
    plt.ylim(-50,450)
    plt.plot(x, sum([coefs[i]*x**i for i in range(n)]), linestyle='-', color='black')

draw_polynomial(coefs)
```

```python
X = []
Y = []
for i in range(40):
    x = random.uniform(-5,5)
    y = polynomial(coefs, x) + random.gauss(0,2)
    X.append(x)
    Y.append(y)
```

```python
plt.scatter(X, Y)
draw_polynomial(coefs)
```

## Training a polynomial regression model

```python
import turicreate as tc
```

```python
data = tc.SFrame({'x':X, 'y':Y})
data
```

### Adding columns to our dataset corresponding to $x^2, x^3, \ldots, x^{200}$

```python
for i in range(2,200):
    string = 'x^'+str(i)
    data[string] = data['x'].apply(lambda x:x**i)
data
```

### 3c) Split the data so that 75% is used for training.

## Splitting data into training and testing

```python
train, test = data.random_split(.75, seed=0)
```

```python
def display_results(model):
    coefs = model.coefficients
    print("Training error (rmse):", model.evaluate(train)['rmse'])
    print("Testing error (rmse):", model.evaluate(test)['rmse'])
    plt.scatter(train['x'], train['y'], marker='o')
    plt.scatter(test['x'], test['y'], marker='^')
    draw_polynomial(coefs['value'])
    plt.show()
    print("Polynomial coefficients")
    print(coefs['name', 'value'])
```

## Training a polynomial regression model with no regularization


```python
model_no_reg = tc.linear_regression.create(
    train, target='y', l1_penalty=0.0, l2_penalty=0.0, verbose=False, validation_set=None)
```

```python
display_results(model_no_reg)
```

## 3d) Training a polynomial regression model with elastic (L1+L2) regularization

```python
model_elastic_reg = tc.linear_regression.create(
    train, target='y', l1_penalty=0.1, l2_penalty=0.1, verbose=False, validation_set=None)
```

```python
display_results(model_elastic_reg)
```

### 3e) The training and testing error of the model with elastic (L1+L2) regularization are 74.68 and 98.24 respectively. Because the training error and the testing error are both high, this model is underfitting.

### 3f)

## Viewing the predictions in the testing set


```python
predictions = test['x', 'y']
predictions['No reg'] = model_no_reg.predict(test)
predictions['Elastic reg'] = model_elastic_reg.predict(test)
```

```python
predictions
```

## 3g) A better optimal model

```python
optimal_model = tc.linear_regression.create(
    train, target='y', l1_penalty=0, l2_penalty=0.1, verbose=False, validation_set=None)
display_results(optimal_model)
```

## Viewing the predictions in the testing set


```python
predictions = test['x', 'y']
predictions['No reg'] = model_no_reg.predict(test)
predictions['Elastic reg'] = model_elastic_reg.predict(test)
predictions['Optimal'] = optimal_model.predict(test)
```

```python
predictions
```
